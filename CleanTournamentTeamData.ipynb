```python
import pandas as pd
```

```python
cleantournamentteamdata_df = pd.read_csv("CSV/Tournament Team Data.csv")
```

```python
cleantournamentteamdata_df.head()
```

```python
cleantournamentteamdata_df = cleantournamentteamdata_df.drop("TEAM.1", axis = 1)
```

```python
cleantournamentteamdata_df.head()
```

```python
cleantournamentteamdata_df.dtypes
```

```python
cleantournamentteamdata_df = cleantournamentteamdata_df.dropna(axis='columns', how='all')
cleantournamentteamdata_df.head()
```

```python
cleantournamentteamdata_df["Win"] = ['1' if i == 1 else '0' for i in cleantournamentteamdata_df["ROUND"]]
cleantournamentteamdata_df.head()
```

```python
cleantournamentteamdata_df.to_csv("CleanTournamentTeamData.csv", index=False)
```
